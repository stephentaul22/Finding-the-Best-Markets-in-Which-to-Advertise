# Finding the Best Markets to Advertise In
Online programming courses are very popular nowadays. Today, I will assume that I work for an online education company, and I will determine which markets are best to advertise in for this company.

Assumptions: Most of my company's courses are on web and mobile development, but we also offer courses in data science, game development, and others. We need to know which 2 markets to begin advertising in.

I will use data from FreeCodeCamp, a popular online programming courses company. The data can be found at their Github repository <a href='https://github.com/freeCodeCamp/2017-new-coder-survey'>here</a>. I will use the clean dataset.

Question:<br/>
Which 2 markets should I begin advertising my company's courses in?

## Setting the Statistical Definitions and Evaluating Representativity
The population relevant to my statistical question is the set of all new programmers that show interest in online programming courses.

Is my sample representative of this population?

```python
fcc['JobRoleInterest'].value_counts(normalize=True).head(150)*100
```

This list is full of new programmers interested in web development, data science, mobile development, game development, and other programming fields. 

Many students show interest in more than one field of study. Some students do not show interest in our primary fields of study, but most show interest in at leaste 1.

Assuming we are focusing intently on our web and mobile development courses, how many students from this list will show interest in our courses then?

```python
import matplotlib.pyplot as plt
%matplotlib inline

ind_red = '#cd5c5c'

top_five_job_roles = fcc['JobRoleInterest'].value_counts(normalize=True).head(5)*100

top_five_job_roles.plot.bar(color = ind_red, figsize=(12,8), fontsize=16, rot=45)
plt.title('Top 5 Job Roles of Interest in the FreeCodeCamp Community', fontsize=16)
plt.ylim([0,14])
plt.tick_params(left=False, bottom=False)
plt.yticks([])
plt.text(-0.15,12.1,str(round(top_five_job_roles[0],1)) + '%', fontsize=18)
plt.text(0.85,6.7,str(round(top_five_job_roles[1],1)) + '%', fontsize=18)
plt.text(1.85,2.5,str(round(top_five_job_roles[2],1)) + '%', fontsize=18)
plt.text(2.85,2.3,str(round(top_five_job_roles[3],1)) + '%', fontsize=18)
plt.text(3.85,2.0,str(round(top_five_job_roles[4],1)) + '%', fontsize=18)

plt.show()
```

```python
fs_web = round(top_five_job_roles[0],1)
fe_web = round(top_five_job_roles[1],1)
be_web = round(top_five_job_roles[3],1)
m_dev = round(top_five_job_roles[4],1)

top_5_pct_web_mobile = fs_web+fe_web+be_web+m_dev
print(round(top_5_pct_web_mobile,1))
```

```python
web_mobile_chart = fcc['JobRoleInterest'].dropna().str.contains('Web Developer|Mobile Developer').value_counts()
nr_web_mobile = web_mobile_chart[True]
nr_not_web_mobile = web_mobile_chart[False]
print(round(nr_web_mobile/(nr_web_mobile+nr_not_web_mobile)*100,2))
```

```python
dark_dark_gray = '#121414'

web_mobile_chart.plot.bar(color = ind_red, figsize=(10,6), fontsize=16, rot=0)
plt.title('Number of Programmers Interested in Web or Mobile Development vs Not Interested in the Same', fontsize=16)
plt.tick_params(left=False, bottom=False)
plt.ylim([0,7000])
plt.yticks([])
plt.text(-0.09,6130,nr_web_mobile,fontsize=20, color=dark_dark_gray)
plt.text(0.94,1062, nr_not_web_mobile, fontsize=20, color=dark_dark_gray)
plt.show()
```

4 of the top 5 most common job roles of interest chosen by new programmers on FreeCodeCamp are either some type of web developer or mobile developer. 

21.9% of FreeCodeCamp programmers who answered this question fall into this section (top 4 of 5 most common job role interest chosen being web development or mobile development specifically).

However, most students chose more than one job role as their job role of interest.

86.2% of students who answered this question chose at least one of our primary fields, mobile development or web development, as their intended job role of interest.

Since our website offers courses primarily in mobile or web development, but also other courses, this seems to be a pretty good set of data so far for our purposes.

```python
fcc_jobrole = fcc[fcc['JobRoleInterest'].notnull()]
```

```python
country_live_absolute = fcc_jobrole['CountryLive'].value_counts()
country_live_relative = fcc_jobrole['CountryLive'].value_counts(normalize=True)*100

pd.DataFrame({'Absolute frequency': country_live_absolute, 'Percentage': country_live_relative})
```

At this point in my analysis, the 2 new markets that we should advertise in would seem to be the United States of America and India. However, there is more data to work with. There is a 'MoneyForLearning' column that can help determine which locations not only show interest, but will be more capable of paying for our courses!

Assuming our company sells subscription based learning at $59 per month, then we are interested in finding learners who can afford that subscription on a monthly basis.

The 'MoneyForLearning' column shows how much money each programmer has spent on their programming education up until the moment of the survey. And the 'MonthsProgramming' column shows how many months the new programmer has been programming.

Dividing these two columns will give us a clue as to which students can afford our courses.

```python
fcc_jobrole['MoneyForLearning']
```

```python
pd.options.mode.chained_assignment = None #Disabling the SettingWithCopy warning since it is benign in this case    

fcc_jobrole['MonthsProgramming'].replace(0,1, inplace = True)

pd.options.mode.chained_assignment = 'warn' #Re-enabling the SettingWithCopy warning
```

```python
fcc_jobrole['MonthsProgramming'].value_counts(dropna=False).sort_index().head(5)
```

```python
pd.options.mode.chained_assignment = None #Disabling the SettingWithCopy warning since it is benign in this case

fcc_jobrole['Monthly'] = fcc_jobrole['MoneyForLearning']/fcc_jobrole['MonthsProgramming'].copy()

pd.options.mode.chained_assignment = 'warn' #Re-enabling the SettingWithCopy warning

fcc_jobrole['Monthly'].isnull().value_counts()
```

```python
fcc_clean = fcc_jobrole[fcc_jobrole['Monthly'].notnull()]
```

```python
fcc_clean['CountryLive'].value_counts().head(10)
```

The data still shows similar proportions in terms of new programmers per country. For the rest of the analysis I will reduce my focus to just the United States, India, the United Kingdom, Canada, and Australia, since these countries all have English as an official language, and I am assuming that our website is written in English. 

```python
country_live_mean = fcc_clean.groupby(['CountryLive']).mean()

countries = ['United States of America', 'India', 'United Kingdom', 'Canada', 'Australia']
country_live_mean['Monthly'][countries].sort_values(ascending=False)
```

```python
exp_money = {}

for country in countries:
    exp_money[country] = fcc_clean['CountryLive'].value_counts()[country]*country_live_mean['Monthly'][country]
    
for val in sorted(exp_money, key=exp_money.get, reverse=True):
    print(val, exp_money[val])
```

The above chart shows that in a perfect world, on average we will earn more money from new programmers in the United States of America and India.

But we also need to check for outliers in the data that may have skewed the averages.

```python
five_countries = fcc_clean[fcc_clean['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada|Australia', na=False)]

fig, ax = plt.subplots(figsize=(14, 8))

import seaborn as sns

sns.boxplot(y = 'Monthly', x = 'CountryLive',
            data = five_countries)
plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Money per Month (USD)', fontsize=16)
plt.xlabel('Country', fontsize=16)
plt.xticks(range(5), ['US', 'UK', 'India', 'Canada','Australia'], fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

Most of the data seems reasonable. There are a couple outliers in the US category, it seems, who reportedly spent more than 50,000 dollars per month. This is extremely unlikely, so I will remove all values greater than $20,000 to be sure enough, and then redo the analysis.

```python
fcc_clean = fcc_clean[fcc_clean['Monthly'] < 20000]
```

```python
country_live_mean = fcc_clean.groupby(['CountryLive']).mean()

countries = ['United States of America', 'India', 'United Kingdom', 'Canada', 'Australia']
country_live_mean['Monthly'][countries].sort_values(ascending=False)
```

```python
five_countries = fcc_clean[fcc_clean['CountryLive'].str.contains(
    'United States of America|India|United Kingdom|Canada|Australia', na=False)]

fig, ax = plt.subplots(figsize=(14, 8))

import seaborn as sns

sns.boxplot(y = 'Monthly', x = 'CountryLive',
            data = five_countries)
plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Money per Month (USD)', fontsize=16)
plt.xlabel('Country', fontsize=16)
plt.xticks(range(5), ['US', 'UK', 'India', 'Canada','Australia'], fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

There still seem to be a few suspicious marks in most countries. Many programmers here have reportedly spent more than $2,500 per month on learning, which is unusual to say the least. It could be that these learners participated in bootcamps, and that could be what causes their monthly cash spent to be so high. I will now investigate to find out which big spenders have participated in boot camps or not, and only keep the ones that have.

## Data Analysis
After cleaning the data of outliers and visualizing it, it is clear one country that we would like to advertise in: the United States. Besides that, it is still uncertain.

Australia has the highest average monthly money spent after the United States, but that number is clearly proped up largely by 3 big spenders. The box plot above shows that if those 3 big spenders are ignored (maybe they entered their data wrong perhaps), then India or Canada would be the next best country for our marketing campaign. However, Australia does have the highest median value, giving it the edge over India and Canada. Finally, it has the leaset amount of programmers in our analysis compared to India and Canada, which does not bode well for it.

India shows a great deal of interest in programming for web development and mobile development, and the learners that we see in this dataset clearly demonstrate a solid average amount of money spent per month of learning programming. The median value spent is zero, which is the same as Canada, which does not look promising. It means that half of its programmers on FreeCodeCamp (228) have spent 0 dollars per month of learning programming. However, India does have the largest pool of programmers among itself, Canada, and Australia, which is a great sign.

Not to be outdone, Canada shows a higher average of money spent per programmer than India. Canada shows fewer programmers in the FreeCodeCamp database than India, but more than Australia. Additionally, a great deal of Canada's average is propped up by one outlier who took a bootcamp course. Nevertheless, it is still a much higher average. It's median value for money spent per month is actually zero though, which means half (120) of its programmers on FreeCodeCamp have spent 0 dollars per month on learning programming.
